```python
import pandas as pd

df = pd.read_csv("vehicles.csv")
df.info()

```

Build custom database for analyst

```python
fuel_economy_study = df[['year', 'make', 'model', 'VClass','cylinders', 'displ', 'highway08', 'comb08',
                          'pv2', 'lv2', 'pv4', 'lv4', 'hpv', 'hlv', 'drive']].copy()
fuel_economy_study = fuel_economy_study.rename(columns={'comb08':'CombCityHwy MPG',
                                                        'hlv':'HatchLuggVol', 'lv2':'2D LuggVol', 'lv4':'4D LuggVol', 
                                                        'pv2': '2D PassVol', 'highway08': 'HW mpg@55', 'pv4':'4D PassVol',
                                                          'hpv':'HatchPassVol'})
fuel_economy_study['HW mpg@65'] = (fuel_economy_study['HW mpg@55']*.9).astype(int)
fuel_economy_study['HW mpg@75'] = (fuel_economy_study['HW mpg@65']*.89).astype(int)
fuel_economy_study['HW mpg@85'] = (fuel_economy_study['HW mpg@75']*.88).astype(int)
fuel_economy_study['HW mpg@95'] = (fuel_economy_study['HW mpg@85']*.87).astype(int)
fuel_economy_study['HW mpg@105'] = (fuel_economy_study['HW mpg@95']*.85).astype(int)
fuel_economy_study['AggCombMPG'] = (fuel_economy_study['CombCityHwy MPG']*.8).astype(int)
fuel_economy_study['CombCHW_Cost100miles'] = round((4/(fuel_economy_study['CombCityHwy MPG'])*100),2)
fuel_economy_study['AggCombCHW_Cost100miles'] = round((4/(fuel_economy_study['AggCombMPG'])*100),2)
fuel_economy_study['HW@55_Cost100miles'] = round((4/(fuel_economy_study['HW mpg@55'])*100),2)
fuel_economy_study['HW@65_Cost100miles'] = round((4/(fuel_economy_study['HW mpg@65'])*100),2)
fuel_economy_study['HW@75_Cost100miles'] = round((4/(fuel_economy_study['HW mpg@75'])*100),2)
fuel_economy_study['HW@85_Cost100miles'] = round((4/(fuel_economy_study['HW mpg@85'])*100),2)
fuel_economy_study = fuel_economy_study[['year', 'make', 'model', 'VClass', 'cylinders', 'displ', 
                           '2D PassVol', '2D LuggVol', '4D PassVol', '4D LuggVol', 'HatchPassVol',
                             'HatchLuggVol', 'drive','CombCityHwy MPG', 'CombCHW_Cost100miles', 'AggCombMPG', 'AggCombCHW_Cost100miles',
                              'HW mpg@55', 'HW@55_Cost100miles','HW mpg@65', 'HW@65_Cost100miles',
                              'HW mpg@75', 'HW@75_Cost100miles', 'HW mpg@85', 'HW@85_Cost100miles', 'HW mpg@95', 'HW mpg@105']]
fuel_economy_study = fuel_economy_study[fuel_economy_study['year'] == 2023]
fuel_economy_study.head()
```

Size of Custom DF

Sort Block  / Drop Unneeded Vehicle Types and Companies

```python
# fuel_economy_study.sort_values(['year', '4D LuggVol', 'HW mpg@55'], ascending=[False, False, False]).head()
# fuel_economy_study.sort_values(['HW mpg@55'], ascending=[True]).head()

# fuel_economy_study[fuel_economy_study['model'].str.contains('Bronco')]
fuel_economy_study.drop(fuel_economy_study[fuel_economy_study['VClass'] == 'Two Seaters'].index, inplace = True)
fuel_economy_study.drop(fuel_economy_study[fuel_economy_study['cylinders'] == 12].index, inplace = True)
fuel_economy_study.drop(fuel_economy_study[fuel_economy_study['make'] == 'Ferrari'].index, inplace = True)
fuel_economy_study.drop(fuel_economy_study[fuel_economy_study['make'] == 'Lamborghini'].index, inplace = True)
fuel_economy_study.drop(fuel_economy_study[fuel_economy_study['make'] == 'Aston Martin'].index, inplace = True)
fuel_economy_study.drop(fuel_economy_study[fuel_economy_study['make'] == 'Maserati'].index, inplace = True)
fuel_economy_study.drop(fuel_economy_study[fuel_economy_study['make'] == 'Bentley'].index, inplace = True)
fuel_economy_study.info()
```

Save to CSV

```python
fuel_economy_study.to_csv('fuel_economy_study.csv', encoding='utf-8')
```
